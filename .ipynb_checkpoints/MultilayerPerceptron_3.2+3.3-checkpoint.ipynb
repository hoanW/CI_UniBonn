```python
#Neural Network's properities: 
#1. Multilayer Perceptron with 2 Inputs + 1 Hidden Layer (20 Unit) + 1 Ouput 
#2. for a nonlinear regression problem
#2. Fermi func als Activationsfunction 
#3. Early Stopping with preset epoch
#4. Learing-Rate = 1

import numpy as np
import matplotlib.pyplot as plt


#fermi function for hidden layer, identity function for output layer
fermi = lambda z: 1/(1+np.exp(-z))    
fermi_deriv = lambda z: np.exp(-z)/((1+np.exp(-z))**2)
identity = lambda z: z
    
class MLP:


    def __init__(self, hiddenlayer_weights, outlayer_weights):
        self.hiddenlayer_weights = hiddenlayer_weights
        self.outlayer_weights = outlayer_weights   
    
    #notice x :inputs
    #       z: weighted sum
    #       a: weighted sum after using activation function

    def feedforward(self, input):
        input = np.array(input)
        
        _, _, hiddenLayer_a = self.compute_layer(fermi, input, self.hiddenlayer_weights)
        _, _, outputLayer_a = self.compute_layer(identity, hiddenLayer_a, self.outlayer_weights)
        return outputLayer_a[0]

    
    def compute_layer(self, activation_func, input, weights): 
        # add bias node
        x = np.insert(input, 0, 1)
        # weighted sum
        z = np.dot(x, weights)
        # weighted sum after using activation function
        a = np.vectorize(activation_func)(z)
        return x, z, a

    def train_outputLayer(self, learning_rate, input, target):
        input = np.array(input)
        # feedforward
        x_with_bias, z_hidden, a_hidden =  self.compute_layer(fermi, input, self.hiddenlayer_weights)
        a_hidden_with_bias, z_output, y = self.compute_layer(identity, a_hidden, self.outlayer_weights)

        # backpropagation
        for i in range(self.outlayer_weights.shape[0]):
            # ∂E / ∂w_i = (∂E / ∂y) * (∂y / ∂z) * (∂z /∂w_i)
            # = (y - t) * 1 * a_i
            
            gradient = (y[0] - target) * a_hidden_with_bias[i]
            self.outlayer_weights[i][0] -= gradient * learning_rate

    def train_hiddenLayer(self, learning_rate, input, target):
        input = np.array(input)

        # feedforward
        x_with_bias, z_hidden, a_hidden = self.compute_layer(fermi, input, self.hiddenlayer_weights)
        a_hidden_with_bias, z_output, y = self.compute_layer(identity, a_hidden, self.outlayer_weights)

        # backpropagation
        for i in range(self.hiddenlayer_weights.shape[0]):
            for j in range(self.hiddenlayer_weights.shape[1]):       
                # ∂E / ∂w_i = (∂E / ∂y) * (∂y / ∂z) * (∂z / ∂a_i) * (∂a_i / ∂z_i) * (∂z_i / w_i)
                #   = (y - t) * 1 * w_i * fermi_deriv(z_i) * input_i
                
                gradient = (y[0] - target) * self.outlayer_weights[j+1][0] *  fermi_deriv(z_hidden[j]) * x_with_bias[i]
                self.hiddenlayer_weights[i][j] -= gradient * learning_rate
    
    def squaredError(self, inputs, target_outputs):
        outputs = np.array([self.feedforward(x) for x in inputs])
        return np.mean((outputs - target_outputs) ** 2)

    
    
if __name__ == '__main__':
    
    #Init Parameter + Inputs/Outputs
    
    inputs = np.linspace(-10, 10 ,1001)
    out_func = lambda x: np.cos(x/2)+np.sin(4/(np.abs(x)+0.3)) - 0.1*x
    outputs_target = np.vectorize(out_func)(inputs)

    learning_rate = 0.05
    n_input = 1
    n_hidden = 20
    n_output = 1
    weights ={
        "hidden" : np.random.uniform(low =-1, high =1, size=(n_input+1,n_hidden)),
        "out" : np.random.uniform(low =-1, high =1, size=(n_hidden+1,n_output))
    }
    
    #Init Graph Plot
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,3,1])
    ax.set(title= 'f(x) = cos(x/2) + sin(4/(|x|+0.3)) – 0.1.x',ylabel='f(x)', xlabel='x')
    ax.set_xlim(-10,10)
    ax.plot(inputs,outputs_target)

    #3_2 : Train in output layer only

    mlp3_2 = MLP(weights["hidden"], weights["out"])
    
    error3_2 = list()
    for epoch in range(100):
        for input, target in zip(inputs, outputs_target):
            mlp3_2.train_outputLayer(learning_rate, input, target)
            
        error = mlp3_2.squaredError(inputs, outputs_target)
        error3_2.append(error)
            
    #3_2 : Outputs Plot
    outputs_3_2 = [mlp3_2.feedforward(x) for x in inputs]
    ax.plot(inputs, outputs_3_2, label="output_3_2")
   
    #3_3: Train in both layers
    mlp3_3 = MLP(weights["hidden"], weights["out"])
    
    error3_3 = list()
    for epoch in range(100):
        for input, target in zip(inputs, outputs_target):
            mlp3_3.train_hiddenLayer(learning_rate, input, target)
            mlp3_3.train_outputLayer(learning_rate, input, target)
            
        error = mlp3_3.squaredError(inputs, outputs_target)
        error3_3.append(error)
        
    #3_3: Outputs Plot
    outputs_3_3 = [mlp3_3.feedforward(x) for x in inputs]
    ax.plot(inputs, outputs_3_3, label="output_3_3")
    ax.legend(loc='best')
    
    
    #Error Plots
    fig2 = plt.figure()
    ax2 = fig2.add_axes([0,0,3,1])
    ax2.set(title= 'Error 3_2',ylabel='error', xlabel='traning index')
    ax2.set_xlim(0,100)
    ax2.plot(error3_2, label = "error_3_2")
    ax2.plot(error3_3, label = "error_3_3")
    ax2.legend(loc='best')
    
    plt.show()
```
